Part 1: Empirical Laws and Curve Fitting
Task 1: Linear Curve Fitting.
Problem:
Fit a straight line y=mx+c to the following data points using the Method of Least Squares:
x=[1,2,3,4,5] y=[3,6,8,11,15]
Required:
1. Compute the slope (m) and intercept (c).
2. Plot the data points and the fitted line on the same graph

```python
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([3, 6, 8, 11, 15])


coefficients1 = np.polyfit(x1, y1, 1)
m1, c1 = coefficients1
fit_line1 = m1 * x1 + c1

plt.scatter(x1, y1, color='red', label='Data Points')
plt.plot(x1, fit_line1, color='blue', label=f'Fit: y = {m1:.2f}x + {c1:.2f}')
plt.title('Linear Fit: y = mx + c')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

Task 2: Polynomial Curve Fitting.
Problem:
Fit a quadratic curve y=ax^2+bx+c to the following data points:
x=[0,1,2,3,4] y=[2,3,6,11,18]
Required:
1. Determine the coefficients a,b,c using the Method of Least Squares.
2. Plot the original data points and the fitted curve on the same graph.

```python
x2 = np.array([0, 1, 2, 3, 4])
y2 = np.array([2, 3, 6, 11, 18])

coefficients2 = np.polyfit(x2, y2, 2)
a2, b2, c2 = coefficients2
fit_curve2 = a2 * x2**2 + b2 * x2 + c2

plt.scatter(x2, y2, color='green', label='Data Points')
plt.plot(x2, fit_curve2, color='purple', label=f'Fit: y = {a2:.2f}x² + {b2:.2f}x + {c2:.2f}')
plt.title('Quadratic Fit: y = ax² + bx + c')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

Task 3: Exponential Curve Fitting.
Problem:
Fit an exponential model y= a* e^(bx) to the following data:
x=[1,2,3,4,5] y=[2.5,4.7,8.8,16.2,30.3]
Required:
1. Estimate a and b using a non-linear fitting technique.
2. Display the original data points and the fitted curve.

```python
x3 = np.array([1, 2, 3, 4, 5])
y3 = np.array([2.5, 4.7, 8.8, 16.2, 30.3])

def exp_model(x, a, b):
    return a * np.exp(b * x)

params3, _ = curve_fit(exp_model, x3, y3)
a3, b3 = params3
fit_curve3 = exp_model(x3, a3, b3)

plt.scatter(x3, y3, color='orange', label='Data Points')
plt.plot(x3, fit_curve3, color='red', label=f'Fit: y = {a3:.2f}e^({b3:.2f}x)')
plt.title('Exponential Fit: y = a * e^(bx)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

Task 4: Laws Containing Three Constants.
Problem:
Fit the model y= a + b*ln(x) + cx^2 to the data:
x=[1,2,3,4,5] y=[2.1,3.6,6.3,11.5,18.9]
Required:
1. Compute the constants a,b,c.
2. Plot the fitted curve and the original data.

```python
x4 = np.array([1, 2, 3, 4, 5])
y4 = np.array([2.1, 3.6, 6.3, 11.5, 18.9])

def three_const_model(x, a, b, c):
    return a + b * np.log(x) + c * x**2
params4, _ = curve_fit(three_const_model, x4, y4)
a4, b4, c4 = params4
fit_curve4 = three_const_model(x4, a4, b4, c4)

plt.scatter(x4, y4, color='blue', label='Data Points')
plt.plot(x4, fit_curve4, color='green', label=f'Fit: y = {a4:.2f} + {b4:.2f}ln(x) + {c4:.2f}x²')
plt.title('Three Constants Model')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

Part 2: Finite Differences
Task 5: Forward Difference Table.
Problem:
Construct a forward difference table for the following data:
x=[0,1,2,3,4] y=[1,3,7,13,21]
Required:
1. Show the forward difference table.
2. Verify that the third-order differences are constant

```python
x5 = np.array([0, 1, 2, 3, 4])
y5 = np.array([1, 3, 7, 13, 21])

n = len(y5)
forward_diff = np.zeros((n, n))
forward_diff[:, 0] = y5

for j in range(1, n):
    for i in range(n - j):
        forward_diff[i, j] = forward_diff[i + 1, j - 1] - forward_diff[i, j - 1]

print("Forward Difference Table:")
print(forward_diff)
```

Task 6: Backward Difference Table.
Problem:
Construct a backward difference table for the following data:
x=[5,6,7,8,9] y=[1,8,27,64,125]
Required:
1. Show the backward difference table.
2. Verify that the second-order differences are constant

```python
x6 = np.array([5, 6, 7, 8, 9])
y6 = np.array([1, 8, 27, 64, 125])

n = len(y6)
backward_diff = np.zeros((n, n))
backward_diff[:, 0] = y6

for j in range(1, n):
    for i in range(j, n):
        backward_diff[i, j] = backward_diff[i, j - 1] - backward_diff[i - 1, j - 1]

print("Backward Difference Table:")
print(backward_diff)
```
